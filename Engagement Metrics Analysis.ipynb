```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Page Views', 'Average Time Spent (mins)', 'Bounce Rate (%)']
values = [10000, 2, 45]

# Bar Graph
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color=['blue', 'orange', 'green'])
plt.title('Engagement Metrics for "Why is the Tower of God Show So Popular?"')
plt.ylabel('Values')
plt.grid(axis='y')

# Display the bar graph
plt.show()
```
